```python
numbers = []
print(numbers)
```

```python
n = [1, 7, 3, 6, 5, 2, 13, 14]

for i in range(len(n)):
    numbers.append(n[i])
print(numbers)
```

```python
i = 0

while i < len(numbers):
    if numbers[i] % 2 == 1:
        del numbers[i]
    else:
        i += 1
print(numbers)
```
